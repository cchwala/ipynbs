### Einkommenssteuer

```python
def einkommenssteuer(brutto):
    if brutto <= 8130:
        return 0
    elif brutto > 8130 and brutto <= 13469:
        y = (brutto - 8130)/1e4
        return (933.7 * y + 1400) * y
    elif brutto > 13469 and brutto <= 52881:
        z = (brutto - 13469)/1e4
        return (228.74 * z + 2397) * z + 1014
    elif brutto > 52881 and brutto <= 250730:
        return 0.42 * brutto - 8196
    elif brutto > 250730:
        return 0.45 * brutto - 15718
    else:
        raise ValueError('brutto must be positive number') 
```

```python
brutto_list = arange(8000,300000)

eks = array([einkommenssteuer(brutto) for brutto in brutto_list])

d_eks = eks[1:-1] - eks[0:-2]
```

```python
figsize(16,5)
plot(brutto_list, 100*eks/brutto_list, label='Effekitv')
plot(brutto_list[0:-2], 100*d_eks, label='Grenzsteuersatz')
xlabel('Bruttoeinkommen in Euro'); ylabel('Steuersatz in %'); legend()
```

### Kalte progression

```python
def plot_KP(brutto_vec=arange(5000,300000,2000), 
            steigerung_vec=arange(0,0.05,0.001), 
            inflation=1.0):
    kp = empty((len(steigerung_vec), len(brutto_vec)))
    
    for i_st, st in enumerate(steigerung_vec):
        for i_br, br in enumerate(brutto_vec):
            br_st = br + br * st
            ne = br - einkommenssteuer(br)
            ne_st = br_st - einkommenssteuer(br_st)
            kp[i_st, i_br] = ne_st/ne - inflation - 1
            
    fig, ax = subplots(figsize=(16,8))
    matplotlib.rcParams.update({'font.size':14})
    lev = [-3,-2,-1,0,1,2,3]
    lev = arange(-5,5,0.5)
    cax = ax.contourf(brutto_vec, 100*steigerung_vec, kp * 100, cmap=cm.RdBu, vmin=-5, vmax=5, levels=lev)
    cb = fig.colorbar(cax)
    cb.set_label('Reale Einkommenssteigerung in %')
    ylabel('as')
    cs = ax.contour(brutto_vec, 100*steigerung_vec, kp * 100, levels=lev, colors='k')
    clabel(cs, fontsize=9, inline=1)
    ax.plot([brutto_vec[0], brutto_vec[-1]], 2*[100*inflation], 'k:')
    
    title('Kalte Progression bei einer Inflationsrate von ' + str(inflation*100) + '%')
    xlabel('Bruttoeinkommen in Euro')
    ylabel('Einkommensteigerung in %')
    #show()
    return fig
```

```python
steigerung = arange(0,0.05,0.001)
brutto = arange(5000,300000,2000)
inflation = 0.015

fig = plot_KP(brutto, steigerung, inflation)
```
