```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import re
```

### Read in data

Data stems from the last pages of this PIK report: http://www.pik-potsdam.de/research/publications/pikreports/.files/pr119.pdf

```python
import pyPdf

filename = '/Users/chwala-c/Dropbox/code/misc_ipynb/data/gwl_liste_2009.pdf'

pdf = pyPdf.PdfFileReader(open(filename, "rb"))
textpage = []
for page in pdf.pages:
    textpage.append(page.extractText())
```

Define parser functions

```python
def get_month(txt):
    # seperate alphanumeric combination of page and month from first characters, e.g. '74Januar'
    #page = re.findall(r"[^\W\d_]+|\d+", txt[0:15])[0]
    month = re.findall(r"[^\W\d_]+|\d+", txt[0:15])[0]
    month = month.lower().capitalize()
    return month

def get_gwl_year_days_for_month(txt_month, year):
    import numpy as np
    import re
    
    gwl_days = re.findall(str(year) + '(.*?)' + str(year),  txt_month)
    if gwl_days:
        return gwl_days
    
gwl_data = []
for txt in textpage:
    month = get_month(txt)
    years = np.arange(1881,2010)
    for year in years:
        gwl_days = get_gwl_year_days_for_month(txt, year)
        if gwl_days:
            gwl_days_list = gwl_days[0].split(' ')
            N_days = len(gwl_days_list) - 1
            for day in range(1,N_days):
                #i1 = (day-1)*5+1
                #i2 = i1+4
                #gwl = gwl_days[0][i1:i2]
                #gwl_data.append((year, month, day, gwl))
                gwl = gwl_days_list[day]
                gwl_data.append((year, month, day, gwl))
```

Parse data to pandas DataFrame

```python
df_gwl = pd.DataFrame({'year'  : [gwl_data[i][0] for i in range(len(gwl_data))],
                       'month' : [gwl_data[i][1] for i in range(len(gwl_data))],
                       'day'   : [gwl_data[i][2] for i in range(len(gwl_data))],
                       'gwl'   : [gwl_data[i][3].strip() for i in range(len(gwl_data))]} )
df_gwl.tail(5)
```

### Define functions to count GWLs

```python
def count_gwl_per_year(df_gwl, gwl_list, month=None, season=None):
    count = []
    for year in years:
        gwl_sum = 0
        
        if month==None and season==None:
            for gwl_str in gwl_list:
                gwl_sum += df_gwl[(df_gwl.year==year)].gwl.str.contains('^' + gwl_str + '$').sum()
        elif season==None:
            for gwl_str in gwl_list:
                gwl_sum += df_gwl[(df_gwl.year==year) &
                                  (df_gwl.month==month)].gwl.str.contains('^' + gwl_str + '$').sum()
        elif season=='winter':
            for gwl_str in gwl_list:
                gwl_sum += df_gwl[(df_gwl.year==year) &
                                  (df_gwl.month=='Januar')].gwl.str.contains('^' + gwl_str + '$').sum()
                gwl_sum += df_gwl[(df_gwl.year==year) &
                                  (df_gwl.month=='Februar')].gwl.str.contains('^' + gwl_str + '$').sum()
                gwl_sum += df_gwl[(df_gwl.year==year-1) &
                                  (df_gwl.month=='Dezember')].gwl.str.contains('^' + gwl_str + '$').sum()
        count.append(gwl_sum)
    return count
```

```python
def running_mean(x,y,N):
    smoothed = np.convolve(y, np.ones((N,))/N, 'valid')
    M = len(x)
    l_diff = M - (max(M,N) - min(M,N) + 1)
    x_shortend = x[np.floor(l_diff/2.0):-np.ceil(l_diff/2.0)]
    return x_shortend, smoothed

def plot_running_mean(x, y, N, *args, **kwargs):
    x_shortend, smoothed = running_mean(x, y, N)
    plt.plot(x_shortend,smoothed, *args, **kwargs)
```

# Data analysis

### Count GWL with north and south component

```python
NORTH_COMPONENT = ('WW', 'NWA', 'NWZ', 'NA', 'NZ', 'HNA', 'HNZ', 'HB', 'TRM', 'NEA', 'NEZ')
SOUTH_COMPONENT = ('SWA', 'SWZ', 'SEA', 'SEZ', 'SA', 'SZ', 'TB', 'TRW')
```

```python
count_N = count_gwl_per_year(df_gwl, NORTH_COMPONENT)
count_S = count_gwl_per_year(df_gwl, SOUTH_COMPONENT)
```

```python
plt.figure(figsize=(16,6))
plt.plot(years, count_N, 'b', label='North component', alpha=0.5)
plt.plot(years, count_S, 'r', label='South component', alpha=0.5)
plot_running_mean(years, count_N, 10, 'b')
plot_running_mean(years, count_S, 10, 'r')
plt.title('GWL whole year'); plt.ylabel('#days')
plt.legend(); plt.show()
```

### Count GWL with north and south component in winter time (DJF)

```python
count_N_DJF = count_gwl_per_year(df_gwl, NORTH_COMPONENT, season='winter')
count_S_DJF = count_gwl_per_year(df_gwl, SOUTH_COMPONENT, season='winter')
```

```python
plt.figure(figsize=(16,6))
plt.plot(years, count_N_DJF, 'b', label='North component', alpha=0.5)
plt.plot(years, count_S_DJF, 'r', label='South component', alpha=0.5)
plot_running_mean(years, count_N_DJF, 10, 'b')
plot_running_mean(years, count_S_DJF, 10, 'r')
plt.title('GWL only DJF'); plt.ylabel('#days')
plt.legend(); plt.show()
```

### Count GWL for potentially cold, warm and snowy winter weather

Accordign to analysis for met station Hoher Peissenberg (Page 33, http://www.pik-potsdam.de/research/publications/pikreports/.files/pr100.pdf)

```python
COLD = ('NA', 'HNA', 'NEA', 'NEZ', 'HFA', 'HFZ', 'HNFA', 'HNFZ')
WARM = ('WZ', 'SWA', 'SWZ', 'SZ', 'TB')
SNOW = ('NWZ', 'NZ', 'NEZ') # Colder and more Precip than average
```

```python
count_COLD_DJF = count_gwl_per_year(df_gwl, COLD, season='winter')
count_WARM_DJF = count_gwl_per_year(df_gwl, WARM, season='winter')
count_SNOW_DJF = count_gwl_per_year(df_gwl, SNOW, season='winter')
```

```python
plt.figure(figsize=(16,6))
plt.plot(years, count_COLD_DJF, 'b', label='Cold', alpha=0.4)
plt.plot(years, count_WARM_DJF, 'r', label='Warm', alpha=0.4)
plt.plot(years, count_SNOW_DJF, 'k', label='Snowy', alpha=0.4)
plot_running_mean(years, count_COLD_DJF, 10, 'b', alpha=0.8)
plot_running_mean(years, count_WARM_DJF, 10, 'r', alpha=0.8)
plot_running_mean(years, count_SNOW_DJF, 10, 'k', alpha=0.8)
plt.title('GWL only DJF'); plt.ylabel('#days')
plt.legend(); plt.show()
```
